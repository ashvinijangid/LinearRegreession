# Importing Libs

```python
import numpy as np
from sklearn.datasets import make_regression
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
plt.style.use("fivethirtyeight")
%matplotlib inline
```

# Generateing Linear Data

```python
x,y = make_regression(100,noise=10,n_features=1)
```

```python
def plot(x,y):
    fig = plt.figure(figsize=(20,5),dpi=150)
    ax = fig.add_subplot(111)
    ax.set_facecolor("white")
    ax.scatter(x,y,c="cyan",edgecolor='k',s=100, alpha=0.7)
    return ax
```

# Plotting the data

```python
plot(x,y)
```

## Linear Regression with Gardient descent

```python
def calculate_weight(x, y):
    print("Iter.", "Error", "\tBias", '\t\tWeights', sep="\t")
    x = np.insert(x, 0, 1, axis=1)
    theta = np.zeros(len(x[0]))
    m = y.size
    lr = 0.1
    for i in range(1,20):
        h = x @ theta
        error = ((y - h)**2).mean()
        theta = theta - lr * 1 / m * (x.T.dot(h - y))

        print(f"{i:<8}{error:^8.4}\t{theta[0]:<8.3}\t\t{theta[1:]}")
#         print(str(i).center(10), error, theta.round(3),
#               sep="\t")
    return theta


def Predict(p, theta):
    x = np.insert([p], 0, 1, axis=0)
    return x.dot(theta)
```

```python
theta = calculate_weight(x, y)
```

```python
x_min,x_max = x.min(),x.max()
ax = plot(x,y)
ax.plot([x_min,x_max],
        [Predict(x_min,theta),
         Predict(x_max,theta)],c="r", label="Regression")
ax.legend(fontsize=20)
```
