```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_regression
```

```python
X,Y = make_regression(n_features=1, 
                      n_samples=100, noise=10)
```

```python
fig = plt.figure(figsize=(20,5),dpi=200)
plt.scatter(X,Y)
```

```python
def regression():
    errors = {}
    for c in np.arange(-1,1,0.01):
        for m in np.arange(50,70,0.1):
            y = (m*X + c).reshape(-1)
            error = ((Y - y)**2).mean()
            errors[error] = c,m
    return errors[min(errors)]
```

```python
bias,weight = regression()
print(bias, weight)
```

```python
fig = plt.figure(figsize=(20,5), dpi=200)
predict = lambda x:weight*x + bias
plt.scatter(X,Y)
plt.plot([min(X),max(X)],
         [predict(min(X)),predict(max(X))],
         linewidth=10,color="r",alpha=0.3)
```
